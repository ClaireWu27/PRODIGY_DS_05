```python
import pandas as pd
# read data
df=pd.read_csv('./Road Accident Data.csv')
df.head()
```

# 1. handle missing value


```python
df.shape[0]
```

```python
df.isna().any()
```

```python
df=df.dropna()
df.shape[0]
```

# 2. remove duplicates


```python
df=df.drop_duplicates()
df.shape[0]
```

```python
df.info()
```

```python
cols=['Day_of_Week','Junction_Control','Junction_Detail','Accident_Severity','Light_Conditions','Local_Authority_(District)','Carriageway_Hazards','Police_Force','Road_Surface_Conditions','Road_Type','Urban_or_Rural_Area','Weather_Conditions','Vehicle_Type']
for col in cols:
    df[col]=df[col].str.strip().str.capitalize()
df.head()
```

```python
for col in cols:
    unique_values = df[col].unique()
    print(f" {col} unique values: {unique_values}")
```

```python
df['Accident_Severity']=df['Accident_Severity'].replace('Fetal', 'Fatal')
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
